```python
%load_ext autoreload
%autoreload 2

import glob
import scipy.signal
import scipy.io

import seaborn as sns
import matplotlib.pyplot as plt

import pickle
import numpy as np 
import json 
import re
import os

from absl import flags
from absl.flags import FLAGS
import sys

from read_emg import EMGDataset

def printProgressBar (iteration, total, prefix = '', suffix = '', decimals = 1, length = 100, fill = '█', printEnd = "\r"):
    """
    Call in a loop to create terminal progress bar
    @params:
        iteration   - Required  : current iteration (Int)
        total       - Required  : total iterations (Int)
        prefix      - Optional  : prefix string (Str)
        suffix      - Optional  : suffix string (Str)
        decimals    - Optional  : positive number of decimals in percent complete (Int)
        length      - Optional  : character length of bar (Int)
        fill        - Optional  : bar fill character (Str)
        printEnd    - Optional  : end character (e.g. "\r", "\r\n") (Str)
    """
    percent = ("{0:." + str(decimals) + "f}").format(100 * (iteration / float(total)))
    filledLength = int(length * iteration // total)
    bar = fill * filledLength + '-' * (length - filledLength)
    print(f'\r{prefix} |{bar}| {percent}% {suffix}', end = printEnd)
    # Print New Line on Complete
    if iteration == total: 
        print()
        
```

```python
# set pathways up:
save_folder = '/oak/stanford/projects/babelfish/magneto/GaddyPaper/processed_data_1000Hz/'
print_every = 50
```

```python
fnames   = ['trainset2.pkl', 'devset2.pkl', 'testset2.pkl']

FLAGS = flags.FLAGS
FLAGS([''])

for i, fname in enumerate(fnames):
    partitions    = np.zeros(3).astype(bool)
    partitions[i] = True
    print('\n', fname)
    
    dataset = EMGDataset(dev = partitions[1], test = partitions[2], returnRaw= True)
    
    with open(fname, 'wb') as f:
        pickle.dump(dataset, f)
```

```python
from scipy.io import savemat
from pathlib import Path
# this step takes a long time...if changes in preprocessing anticipated will be good to parallelize


FLAGS = flags.FLAGS
FLAGS([''])


with open('trainset2.pkl', 'rb') as f:
    trainset = pickle.load(f)
    
with open('devset2.pkl', 'rb') as f:
    devset = pickle.load(f)
    
with open('testset2.pkl', 'rb') as f:
    testset = pickle.load(f)

print('Partitions loaded. Generating .mat files...')
for partition, folder in zip([trainset, devset, testset], ['train', 'dev', 'test']):
    i = 0
    print(f'======={folder}========')
    
    for x in partition:
        # generate file path
        session     = str(x['session_ids'][0].item())
        file_id     = str(x['file_label'])
        save_dir    = os.path.join(save_folder, folder, f'session_{session}')
        save_fname  = f'{save_dir}/{file_id}.mat'
        Path(save_dir).mkdir(parents=True, exist_ok=True)
        
        if not os.path.isfile(save_fname):
        
            # convert back to numpy so we can save as .mat files
            for key, val in x.items():
                try:
                    x[key] = val.numpy()
                except:
                    pass

            if i % print_every:
                printProgressBar(i, partition.__len__())
            savemat(save_fname, x)
        
        i += 1  
        del x
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

means = list()
SDs   = list()
for i in range(7):
    normalizer = pkl.load(open(glob.glob(save_folder + f'*session{i}_normalizer*')[0], 'rb'))
    means.append(normalizer.feature_means)
    SDs.append(normalizer.feature_stddevs)
    
f, ax = plt.subplots(figsize=(15, 5))


plt.subplot(1, 2, 1)
sns.barplot(data = np.concatenate(means), alpha = 0.6)
for i in range(8):
    plt.scatter(np.ones(7) * i, np.concatenate(means)[:, i])
plt.xlabel('Channel')
plt.ylabel('Mean')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.subplot(1, 2, 2)
sns.barplot(data = np.concatenate(SDs), alpha = 0.6)
for i in range(8):
    plt.scatter(np.ones(7) * i, np.concatenate(SDs)[:, i])
plt.xlabel('Channel')
plt.ylabel('SD')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.suptitle('Multisession channel statistics')
```

```python
emglens  = list()
phonemes = list()

for i in range(7):
    mat_files = glob.glob(save_folder + f'*session{i}*')
    
    session_lens     = list()
    session_phonemes = list()
    for f in mat_files:
        dat    = scipy.io.loadmat(f)
        session_lens.append(dat['emg'].shape[0] / 800)
        session_phonemes.append(dat['phonemes'])
            
    emglens.append(session_lens)
    phonemes.append(session_phonemes)
        
    print(f'Session {i} finished.')
```

```python
# plot some examples

f, ax = plt.subplots(figsize=(20, 10))

examples = np.random.choice(glob.glob(save_folder + '/*/*/*'), 4)

for i, example in enumerate(examples):
    plt.subplot(4, 1, i + 1)
    example = scipy.io.loadmat(example) 
    plt.plot(example['raw_emg'])
    plt.title(example['text'])
plt.show()

for example in examples:
    dat = scipy.io.loadmat(example)
    print('\n')
    print('Text: ', dat['text'])
    #print('Phonemes: ', dat['phonemes'])
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 6))

# Draw a violinplot with a narrower bandwidth than the default
sns.violinplot(data=emglens, palette="Set3", bw=.2, cut=1, linewidth=1)

# Finalize the figure
plt.ylabel('Time (sec)')
plt.xlabel('Session ID')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
import pandas as pd

phonemes_df = list()
for i, x in enumerate(phonemes):
    
    phonemes_df.append(pd.DataFrame({
        'phonemes' : np.concatenate(x),
        'session'  : [i] * len(np.concatenate(x))
    }))
    
phonemes_df = pd.concat(phonemes_df)

f, ax = plt.subplots(figsize=(15, 6))

classes, counts = np.unique(phonemes_df['phonemes'], return_counts= True)

order = np.argsort(counts)[::-1]
classes = classes[order]
counts  = counts[order]

plt.bar(classes, counts)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_yscale('log')
plt.xlabel('Phonemes')
plt.ylabel('Counts (log scale)')
plt.show()
```
